# 5. El Analizador Virtual (Validación de Malus)
Este ejercicio cierra el lazo con tu investigación sobre polarización y la Ley de Malus.  El Reto: Pega un marcador ArUco en el soporte giratorio de tu polarizador analizado. Usa la rotación del marcador para medir el ángulo $\theta$ en el software y compáralo con el ángulo que crees tener físicamente.Por qué importa: Validarás si tu sistema de control tiene errores mecánicos o de alineación (backlash) comparando una medida óptica (ArUco) con una mecánica.
