# <font color = "#EBF527"> Calculadora de propinas color</font>
 ### Logica: Para calcular la propina fácilmente, multiplica el total de la cuenta por el porcentaje deseado en formato decimal 
$$
Propina = total_{   cuenta} \times \%_{deseado}
$$

**Extra:**
En México, la tasa general del Impuesto al Valor Agregado (IVA) es del 16% sobre el valor de la mayoría de los bienes y servicios. 
Por lo que tenemos que considerar  este porcentaje para el total de la cuenta 


```python
def calc_propina():
    Total = float(input("Introduce la cuenta"))
    porcentaje = float(input("Introduce ahora la propina deseada"))/100

    # --- Calcular propina ---
    monto_propina = Total * porcentaje

    # --- Con IVA ---
    monto_iva =  Total * 0.16
    subtotal_con_iva = Total + monto_iva

    # --- Tontal con propina e iba
    total_final = subtotal_con_iva + monto_propina

    print("-"*30)
    print(f"Cuenta Base:     ${Total:>.2f}")
    print(f"IVA (16%):       ${monto_iva:>.2f}")
    print(f"Propina ({porcentaje*100}%): ${monto_propina:>.2f}")
    print(f"TOTAL A PAGAR:   ${total_final:>.2f}")

calc_propina()
```

## <font color = "#EBF527"> Convertidor de unidades temperatura </font>

Farenheit a Celsius: 
$$
Celsius = (Farenheit - 32)/1.8
$$

Celsius a Kelvin:
$$
Kelvin  =  Celsius + 273.15
$$

```python
def conv_Termic_unit():
    '''
    Conversor de unidades termicas 
    Farenheit -> Celsius Kelvin 
    '''
    Farenheit = float(input("Introduce la temperatura °F"))

    Celsius = (Farenheit - 32)/1.8

    # --- Kelvin ---
    Kelvin = Celsius + 273.15

    print("Temperaturas")
    print("-"*30)
    print(f"Farenheit: {Farenheit} °F")
    print(f"Celsius: {Celsius:.2f} °C")
    print(f"Kelvin: {Kelvin:.2
    f} K")

conv_Termic_unit()
```

```python
from tkinter import * #Importamos todas las clases
#Instanciamos el objeto tipo TK
ventana = Tk()

#Metodo absoluto

# --- Ventana ---
ventana.resizable(1,1)
ventana.geometry("200x200")
ventana.title("Ejemplo place")
path_icon = "pacocris.ico"
ventana.iconbitmap(path_icon)


# --- Comandos/Funciones
def fnSuma():
    n1 = txt1.get() #E metodo get nos da el valor de las cajas de texto
    n2 = txt2.get()

    r = float(n1) + float(n2)
    txt3.delete(0,'end')#Borramos el contenido de la caja de texto
    txt3.insert(0,r) #0: Posicion INICIAL DE la caja de texto


# --- Boton ---
btn1 = Button(ventana,text="Sumar",command=fnSuma)
btn1.place(x=230,y=50,width=80,height=30)

# --- Label ---
lbl1 = Label(ventana,text = "Primer numero", bg = "yellow")
lbl1.place(x=10,y=10,width=100,height=30)

lbl2 = Label(ventana,text = "Segundo numero", bg = "yellow")
lbl2.place(x=10,y=50,width=100,height=30)

lbl3 = Label(ventana,text = "Resultado", bg = "yellow")
lbl3.place(x=10,y=120,width=100,height=30)


# --- Cajas de texto --- 
#Entry es una caja de texto donde ingresar cosas
txt1 = Entry(ventana, bg = "pink")
txt1.place(x=120,y=10,width=100,height=30)

txt2 = Entry(ventana, bg = "pink")
txt2.place(x=120,y=50,width=100,height=30)

txt3 = Entry(ventana, bg = "pink")
txt3.place(x=120,y=120,width=100,height=30)


ventana.mainloop()
```
